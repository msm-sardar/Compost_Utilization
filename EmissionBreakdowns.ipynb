```python
from project_class import *
from building_matrices import *
from Compost_use import *
from brightway2 import *
from time import time
from SWOLF_method import *
from AbioticDepletion import import_aibiotic_depletion

Treatment_processes = {}
Treatment_processes['Compost_use']={'input_type':[],'model': Compost_use()}

Treatment_processes['Compost_use']['model'].Compost_input.operation['choice_BU']['amount']=1

demo = project("Compost_use",Treatment_processes)
demo.init_project('SWOLF_AccountMode_LCI DATA.csv')
demo.write_project()
demo.group_exchanges()
import_methods()
import_aibiotic_depletion()
from Import_IPCC_new import *


scenario1 = {"Compost_use":{"Compost":1}}
demo.process_start_scenario(scenario1,'scenario1')
```

```python
db = Database("waste")
functional_unit = {db.get("scenario1") : 1}
```

```python
method = ('IPCC 2013, Ecoinvent V3.5', 'climate change', 'GWP 100a, bioCO2=1, C1_36')
A=LCA(functional_unit,method)
A.lci()
A.lcia()
print(A.score)      
```

```python
top_em = pd.DataFrame(columns=['flow','category','val'])
cutoff = 9
for i,item in enumerate(A.top_emissions()):
    if i < cutoff:
        top_em.loc[i] = [item[2].as_dict()['name'],item[2].as_dict()['categories'],item[0]]

top_em.loc[cutoff+1] = ['Other', '_', A.score - sum(top_em['val'][0:cutoff])]
top_em.loc[cutoff+2] = ['Total', '_', A.score]
top_em
```

```python
top_act = pd.DataFrame(columns=['flow','val'])
cutoff = 6
for i,item in enumerate(A.top_activities()):
    if i < cutoff:
        top_act.loc[i] = [item[2].as_dict()['name'],item[0]]

top_act.loc[cutoff+1] = ['Other', A.score - sum(top_act['val'][0:cutoff])]
top_act.loc[cutoff+2] = ['Total', A.score]
top_act
```
