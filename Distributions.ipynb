```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
Both=pd.read_csv('Both.csv')
ADC=pd.read_csv('ADC.csv')
Fer=pd.read_csv('Fert.csv')
Peat=pd.read_csv('Peat.csv')
```

```python
columns=Both.columns[1:9]
columns
```

# ('IPCC 2013, Ecoinvent V3.5', 'climate change', 'GWP 100a, bioCO2=1, C1_36')

```python
i=0
ylab = 'GWP ('+'$kg\ CO_2 e$'+')'
DF = pd.DataFrame()
DF['ADC'] = ADC[columns[i]]
DF['Soil amendment \n fertilizer offset'] = Fer[columns[i]]
DF['Soil amendment \n peat offset'] = Peat[columns[i]]
DF['Soil amendment \n both fertilizer \n and peat offsets'] = Both[columns[i]]
```

```python
plt.figure(figsize=(12,5))
sns.set(font_scale=1.5) 
sns.set_style("white")
ax=sns.boxplot(data=DF)
ax.set_ylabel(ylab,color='black')
ax.set_xticklabels(DF.columns,color='black')
```
