```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Data_corr.csv')
data.head()
```

```python
plt.rcParams['font.size'] = 13
Fig=plt.Figure(figsize=(10,7), dpi =330)
subplots=Fig.subplots(nrows=1,ncols=5, sharey=True, gridspec_kw = {'wspace':0.15, 'hspace':0})
```

```python
sns.barplot(data=data, x = 'GWP', y='Index', ax=subplots[0], color='black')
sns.barplot(data=data, x = 'Acidification', y='Index', ax=subplots[1], color='black')
sns.barplot(data=data, x = 'Eutrophication', y='Index', ax=subplots[2], color='black')
sns.barplot(data=data, x = 'CED', y='Index', ax=subplots[3], color='black')
sns.barplot(data=data, x = 'ADP', y='Index', ax=subplots[4], color='black')

for i, ax in enumerate(subplots):
    ax.set_xlim(-1,1)
    ax.set_ylabel(None)

subplots[0].set_xlabel(r'$\Delta$ GWP')
subplots[1].set_xlabel(r'$\Delta$ Acidification') 
subplots[2].set_xlabel(r'$\Delta$ Eutrophication') 
subplots[3].set_xlabel(r'$\Delta$ CED') 
subplots[4].set_xlabel(r'$\Delta$ ADP') 
Fig 
```
